```python
def scc(graph):
    # Run Depth First Search to find out the finish time to find each vertex
    marked, sorted_vertices = set(), []
    for k in list(graph.keys()):
        if k not in marked:
            sorted_vertices.extend([v for v in dfs(graph, k, marked)])

    reversed_graph = defaultdict(set)
    while len(graph) > 0:
        vertex, edges = graph.popitem()
        for edge in edges:
            reversed_graph[edge].add(vertex)
    
    marked, scc = set(), defaultdict(set)
    for v in reversed(sorted_vertices):
        if v not in marked:
            scc[v].update([node for node in dfs(reversed_graph, v, marked)])
    return scc
            
def dfs(graph, source, marked):
    """
    Iterative DFS
    """
    stack, popped = [source], set()
    while stack:
        vertex = stack[-1]
        to_visit = graph[vertex] - marked
        if vertex not in marked:
            marked.add(vertex)
            stack.extend(to_visit)
        if not to_visit:
            stack.pop()
            if vertex not in popped:
                popped.add(vertex)
                yield vertex
```

```python
G = defaultdict(set)
# Load data into the graph data structure
with open("wiki-Vote.txt", "r") as ins:
    for line in ins:
        if not line.startswith("#"):
            vertex, edge = [int(i) for i in line.split('\t')]
            G[vertex].add(edge)
```

```python
components = scc(G).values()
giant = max(components, key=lambda x: len(x))
```
